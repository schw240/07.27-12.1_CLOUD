```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
# Tree    2   .
X = iris.data[:,2:]
```

```python
y = iris.target
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier()
```

```python
tree
```

```python
tree.fit(X,y)
```
