```python
from collections import deque

matrix = [[3,2,3,2],
          [2,1,1,2],
          [1,1,2,1],
          [4,1,1,1]]

dx = [-1,0,1,0]
dy = [0, 1, 0, -1]

dis = [[0]*4 for _ in range(4)]

def max_link_cnt(row, col, board, color):
    global dis, dx, dy
    Q = deque()
    Q.append((row, col))
    board[row][col] = 0 #(, ) 0 .
    while Q:
        print(Q)
        tmp = Q.popleft()
        for i in range(4):
            x = tmp[0] + dx[i]
            y = tmp[1] + dy[i]
            #  matrix color  
            if 0 <= x <= 3 and 0<= y <= 3 and board[x][y] == color:
                board[x][y] = 0
                dis[x][y] = dis[tmp[0]][tmp[1]] + 1
                Q.append((x,y))
    # dis   
    digit = [0]*16
    for xx in dis:
        for i in range(1,16):
            digit[i] += xx.count(i)
    #print(digit)
    if digit.count(0) == 16: return -1 #  
    link_cnt = []
    for i in range(sorted(digit, reverse=True)[0]):
        cnt = 0
        for j in range(i+1,16):
            if digit[j]!= 0:
                digit[j] = -1
                cnt += 1
        link_cnt.append(cnt+1)
    print(link_cnt)
    return sorted(link_cnt, reverse=True)[0]

max_link_cnt(2,0, matrix,1)                
```
